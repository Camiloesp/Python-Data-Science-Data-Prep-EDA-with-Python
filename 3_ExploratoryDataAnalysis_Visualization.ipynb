```python
 import pandas as pd
```

```python
groceries = pd.read_pickle('Data/groceries_with_new_columns.pkl')
groceries.head()
```

```python
groceries.loc[groceries['Low Inventory'] == 'Low Inventory']
```

```python
groceries.loc[groceries['Low Inventory'] == 'Low Inventory', ['Price_Dollars', 'Inventory']]
```

```python
groceries.Subcategory[0] # Space before subcategory.
```

```python
groceries[groceries.Subcategory == ' Snacks'] # Add space
```

```python
# Clean spaces in subcategory
groceries.Subcategory = groceries.Subcategory.str.strip()
```

```python
groceries.Subcategory[0]
```

```python
groceries[groceries.Subcategory == 'Snacks']
```

```python
mask = (groceries['Low Inventory'] == 'Low Inventory') & (groceries.Subcategory == 'Snacks')
```

```python
groceries[mask] # Filter rows
```

```python
groceries.loc[mask, ['Subcategory', 'Item', 'Inventory']] # Filter on columns as well
```
