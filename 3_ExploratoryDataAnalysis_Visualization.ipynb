## Filter DataFrame

```python
 import pandas as pd
```

```python
groceries = pd.read_pickle('Data/groceries_with_new_columns.pkl')
groceries.head()
```

```python
groceries.loc[groceries['Low Inventory'] == 'Low Inventory']
```

```python
groceries.loc[groceries['Low Inventory'] == 'Low Inventory', ['Price_Dollars', 'Inventory']]
```

```python
groceries[groceries.Subcategory == ' Snacks'] # Returns nothing but we do have products with the Snacks subcategory.
```

```python
groceries.Subcategory[0] # Space before subcategory.
```

```python
groceries[groceries.Subcategory == ' Snacks'] # Add space
```

```python
# Clean spaces in subcategory
groceries.Subcategory = groceries.Subcategory.str.strip()
```

```python
groceries.Subcategory[0]
```

```python
groceries[groceries.Subcategory == 'Snacks']
```

```python
mask = (groceries['Low Inventory'] == 'Low Inventory') & (groceries.Subcategory == 'Snacks')
```

```python
groceries[mask] # Filter rows
```

```python
groceries.loc[mask, ['Subcategory', 'Item', 'Inventory']] # Filter on columns as well
```

## Sort dataframe

```python
groceries.sort_values('Price_Dollars') # Lowest to highest
```

```python
groceries.sort_values('Price_Dollars', ascending=False) # highest to lowest
```

```python
groceries.sort_values(['Subcategory', 'Price_Dollars'])
```

```python
groceries[groceries['Low Inventory'] == 'Low Inventory'].sort_values('Price_Dollars')
```

## Grouping

```python
groceries.head(3)
```

```python
groceries.groupby('Category')['Inventory'].sum()
```

```python
groceries.groupby('Category')['Inventory'].count()
```

```python
groceries.groupby(['Category', 'Subcategory'])['Inventory'].count().reset_index()
```

```python
(groceries.groupby(['Category', 'Subcategory'])['Inventory']
          .agg(['sum', 'count'])
          .reset_index())
```
